**Théorème :** *Soit $f$ une fonction continue de  $\mathbb{R}^d$ dans $\mathbb{R}^d$.On suppose que:*



***H1:*** Hypothèse sur la fonction $f$: *On suppose qu'il existe $x'\in\mathbb{R}^d$ tel que $f(x')=0$, et que $\langle f(x),x-x'\rangle>0$ pour tout $x\in \mathbb{R}^d$*



***H2:*** Hypothèse sur le pas $\gamma$: $(\gamma_n)_{n\geq1}$ *est une suite décroissante de réels positifs tels que* :$\sum_{n\geq1}\gamma_n<\infty$


***H3*** Hypothèse sur la suite $Y=(Y_n)_{n\geq0}$:*Pour $(\mathcal{F}_n)_{n\geq0}$ une filtration sur un espace de probabilité, et $(Y_n)_{n\geq0}$ une suite de variables aléatoires sur le même éspace de probabilité telles que:*

  +++ ***H3.1:*** $\mathbb{E}\left[Y_{n+1}\mid\mathcal{F}_n\right]=f(X_n)$

  +++ ***H3.2:*** $\mathbb{E}\left[(Y_{n+1}-f(X_n))^2\mid\mathcal{F}_n\right]\leq \sigma^2(X_n)$ , *Ou:*
$$\sigma^2(x)+f^2(x)\leq K(1+\mid x\mid)^2$$

   *On defini alors la suite $(X_n)_{n\geq0}$ par: $X_0=x_0 $, ou $x_0\in\mathbb{R}^d$, et $\forall n\geq 0$*
$$X_{n+1}=X_n-\gamma_nY_{n+1}$$



***Sous ses hypothèses, la suite de variables aléatoires $(X_n)_{n\geq0}$ converge prêsque surement vers $x'$.***



**Remarque:** L'une des utilisation principales de l'algorithma de Robbins-Monro est quand une fonction continue $f$, peut être écrite sous la forme:
$$\forall x \in\mathbb{R} , f(x)=\mathbb{E}\left[F(x,U)\right]$$

Avec $U$ une variable aléatoire de loi connue, $F$ une fonction. Dans ce cas $Y=(Yn)_{n\geq0}$ est définie de la manière suivante:
$$Y_n=F(X_n,U_{n+1})$$

 Ou $(U_n)_{n\geq1}$ est une suite de variables aléatoires indépendantes et indentiquement 

distribuées selon la loi de $U$.

On peut voir alors que si $\mathcal{F}_n=\sigma(U_1,U_2,..,U_n)$,
$$\mathbb{E}\left[F(X_n,U_{n+1})\mid\mathcal{F}_n\right]=f(X_n)$$
En plus 
$$\mathbb{E}\left[\left(F(X_n,U_{n+1})-f(X_n)\right)^2\mid\mathcal{F}_n\right]=\sigma^2(X_n)$$
Ou $\sigma^2(x)=\mathbb{E}\left[\left(F(x,U)-f(x)\right)^2\right]$.

L'hypothèse  ***H3.2*** porte alors sur le comportement de l'espérance et de la variance de $F(x,U)$

# La loi forte des grands nombres comme algorithme stochastique}
On considère $(G_n,n\geq 0)$ une suite de gaussiennes centrées réduites
indépendantes. On pose $M_n=(G_1+\cdots+G_n)/n$



**Proposition:** *Pour tout $n\in\mathbb{N}$*
   $$
      M_{n+1} = M_n -\gamma_n \left(M_n - G_{n+1}\right),
   $$
   où $\gamma_n=1/(n+1)$. 

   En effet,
   
   $$\begin{align*}
   M_{n+1}-M_n&=\frac{n\sum_{i=1}^{n+1}G_i-(n+1)\sum_{i=1}^nG_i}{n(n+1)}\\
   &=\frac{G_{n+1}}{n+1}-\frac{M_n}{n+1}\\
   &=-\gamma_n(M_n-G_{n+1})\\
   \end{align*}$$
  
  
**Théorème:**(Loi des grands nombres pour les martingale) 
  
  *Soit $(S_n)_{n\geq0}$ une martingale de carrée integrable, on a alors:*
  
  +*Sur* $\langle S\rangle_\infty:=\lim_{n\rightarrow \infty}\langle S\rangle_n<\infty$ $S_n$ *converge présque surement vers une variable :$S_\infty$*
  
  +*Sur* $\{\langle S\rangle_\infty=\infty\}$,
  $$\lim_{n\rightarrow\infty}\frac{S_n}{\langle S\rangle_n}=0 \text{ , p.s}$$
   *En plus, si $a(t)$ est u,ne fonction positive, croissante telle que $\int_0^{\infty}\frac{dt}{1+a(t)}<\infty$, alors*
   $$\lim_{n\rightarrow\infty}\frac{S_n}{\sqrt{a(\langle S\rangle_n)}}=0 \text{ , p.s}$$
   
   
   **Proposition:** *Avec les même notations, on pose $S_0=0$ et $S_n=nM_n, \forall n >0$, alors $(S_n)_{n\geq0}$ est une martingale,et $\langle S\rangle_n=n$*
   
   On a 
   $$\begin{align*}
       \mathbb{E}\left\{S_{n+1}\mid\mathcal{F}_n\right\}&=\mathbb{E}\left\{G_1+...+G_n+G_{n+1}\mid\mathcal{F}_n\right\}\\
       &=G_1+...+G_n+0=S_n
   \end{align*}$$
   Donc $(S_n)_{n\geq0}$ est une martingale, en plus $\langle S\rangle_0=0$, et $\forall n>0$:
   $$\langle S\rangle_{n+1}-\langle S\rangle_n=\mathbb{E}\left\{(S_{n+1}-S_n)^2\mid\mathcal{F}_n\right\}=Var\{G_{n+1}\}=1$$
   Donc
   $$\langle S\rangle_n=\sum_{k=0}^{n-1}\langle S\rangle_{k+1}-\langle S\rangle_k=n$$
   
   
   Donc $\langle S\rangle_\infty=\infty$, et en utilisant le théorème précédant on retrouve la loi forte des grands nombre. 
   
   
   
**Remarque:**   Pour avoir une idéé sur la vitésse de convergence, on utilise la fonction $a(t)=Ct^{2-\beta}$, pour $\frac{1}{2}<\beta<1$, on trouve alors en faisant les changements de variables $k=2-\beta$ et dans l'integration $u=(C^{\frac{1}{k}}t)$ et $x=\frac{1}{1+u^k}$  que:
   
   $$\begin{align*}
       \int_0^\infty\frac{dt}{1+a(t)}&=C^{\frac{-1}{k}}\int_0^\infty\frac{du}{1+u^k}\\
       &=\frac{C^{\frac{-1}{k}}}{k}\int_0^1x^{(1-\frac{1}{k})-1}(1-x)^{\frac{1}{k}-1}dx\\
       &=\frac{C^{\frac{-1}{k}}}{k}\beta eta\left(\frac{1}{k},1-\frac{1}{k}\right)<\infty
   \end{align*}$$
   On a alors avec $\gamma_n=\frac{C}{n^\beta}$
   $$\lim_{n\rightarrow\infty}\frac{M_n}{\sqrt{\gamma_n}}=\lim_{n\rightarrow\infty}\frac{S_n}{\sqrt{a(n)}}=\lim_{n\rightarrow\infty}\frac{S_n}{\sqrt{a(\langle S \rangle_n)}}=0 \text{ , p.s}$$

  

```python
import numpy as np;
import math;
from math import exp
import scipy.stats as stats;
import random;
import matplotlib.pyplot as plt;

def gamma(n,C,beta):
    return C / ((n+1)**beta);

def algorithme_rm(x_0,n,C,beta,F,G):
    X=np.zeros(n)
    X[0]=x_0
    for k in range(n-1):
        X[k+1]=X[k]-gamma(k,C,beta)*F(X[k],G[k+1])
    return X




def algorithme_rm2(x_0,n,C,beta,F,G):
    X=np.zeros(n)
    X[0]=x_0
    for k in range(n-1):
        X[k+1]=X[k]-gamma(k,C,beta)**2*F(X[k],G[k+1])
    return X
```

```python
# Dans le cas de la loi forte, F(x,g)= (x-g) et f(x)=x-E(g)=x.
def F1(x,g):
        return x-g

n=10000;

# Beaucoup de variance, la condition initiale est vite oubliee 
C=1
beta=0.75
x_0=1
G=np.random.normal(size=n)
X=algorithme_rm(x_0,n,C,beta,F1,G)
print(X)
plt.plot(X)
plt.plot(np.zeros(n));plt.ylim(-0.25,+1);plt.show()
```

```python
def algorithme_rm_hyper(x_0,n,h,F,G):
    X=np.zeros(n)
    X[0]=x_0
    for k in range(n-1):
        X[k+1]=X[k]-h*F(X[k],G[k+1])#gamma(k,C,beta)*F(X[k],G[k+1])
    return X
```

**Remarque:** Pou obtenir Un théorème cental limite on aura besoin du théorème suivant\\
   
   **Théorème** (TCL pour les martingales)
   *Soient $(S_n)_{n\geq 0}$ une martingale locale de carée integrable et $a(n)$ uen suite croissante de rééls positives qui tends vers $+\infty$. on suppose que :*
   
   **+Condition du crochet:** 
   $$\lim_{n\rightarrow\infty}\frac{1}{a(n)}\langle S\rangle_n=\sigma^2 \text{ , en probabilité}$$\\
   
   **+Condition de Lindeberg:** $\forall\epsilon>0$
   $$
   \lim_{n\rightarrow\infty}\frac{1}{a(n)}\sum_{k=1}^n\mathbb{E}\left\{(\Delta S_{k+1})^2\mathbf{1}_{\{\mid\Delta S_{k+1}\mid\geq\epsilon\sqrt{a(n)}\}}\mid \mathcal{F}_k\right\}=0
   $$
   *Alors:*
   $$\frac{S_n}{\sqrt{a(n)}}\text{ converge en loi vers }\sigma \mathcal{N}(0,1)$$
   
   
   **Remarque:**  On peut voir en guardant les mêmes notation que
   $$\frac{M_n}{\sqrt{\gamma_n}}=\frac{S_n}{\sqrt{a(n)}}$$
   En plus,$\forall n\geq0$ $\frac{1}{C}n^{-\frac{1}{2}}\leq \frac{1}{C}n^{\beta-1}\leq \frac{1}{C}$ il existe alors $\sigma^2\in[0,\frac{1}{C}]$ tel que 
   $$
   \lim_{n\rightarrow\infty}\frac{1}{a(n)}\langle S\rangle_n=\sigma^2
 $$
 Ce qui verifie la condition du crochet, pour la condition de Lindeberg on a pour $\epsilon>0$
 $$\frac{1}{a(n)}\sum_{k=1}^n\mathbb{E}\left\{ G^2_{k+1}\mathbf{1}_{\{\mid G_{k+1}\mid\geq\epsilon\sqrt{a(n)}\}}\mid \mathcal{F}_k\right\}=\frac{n}{a(n)}\mathbb{E}\left\{G^2\mathbf{1}_{\{\mid G\mid\geq\epsilon\sqrt{a(n)}\}}\right\}$$
 La condition de Lindeberg est verifiée grace a l'integrabilité de $G^2$ et au théorème de Lebesque.

Pour $p$ un seuil compris entre $0$ et $1$, on cherche à
calculer la valeur d'un fractile $x_p$ d'ordre $p$ d'une
gaussienne centrée réduite $G$. 

\vspace{2mm}

**Proposition:** $x_p$ *est solution
d'une équation du type*
$$
f(x_p) =\mathbb{E}\left(F(x_p,G)\right)=0,
$$
*Où* $F(x,g)=1_{\{g\leq x\}}-p$ 
\begin{proof}


On a
$$\mathbb{P}(G\leq x_p)=p$$
Donc
$$f(x_p)=\mathbb{E}[1_{\{G\leq x_p\}}-p]=0$$

On prend alors,
$$F(x_p,G)=1_{\{G\leq x_p\}}-p$$

Ceci suggère d'utiliser l'algorithme de Robbins et Monro suivant:
  $$
  X_{n+1}=X_n - \gamma_n F(X_n,G_{n+1}),
  $$   
  où $(G_n,n\geq 1)$ est une suite de variables aléatoires gaussiennes
  centrées réduites et $\gamma_n=C/n^\beta$, où $0<\beta\leq 1$.

  $x_p \approx 1.65$).

  \mathbb{E}\left\{e^{-\lambda G - \frac{\lambda^2}{2}} f(G+\lambda)\right\}
    = \mathbb{E}^{\mathbb{Q}}\left\{f(G+\lambda)\right\}=\mathbb{E}\left\{f(G)\right\},
$$
On cherche a minimiser en $\lambda$ la variance de 
$
   X_\lambda = e^{-\lambda G - \frac{\lambda^2}{2}} f(G+\lambda).
$



**Proposition :**
$$
Var\left\{X_\lambda\right\} 
  = \mathbb{E}\left\{e^{-\lambda G + \frac{\lambda^2}{2}} f^2(G)\right\} - \mathbb{E}\{f(G)\}^2.
$$




On a
$$\begin{align*}
    Var\{X_{\lambda}\}&=\mathbb{E}\left\{(e^{-\lambda G - \frac{\lambda^2}{2}} f(G+\lambda))^2\right\}-\mathbb{E}\{f(G)\}^2\\
    &=\mathbb{E}\left\{e^{-\lambda G - \frac{\lambda^2}{2}} g(G+\lambda)\right\}-\mathbb{E}\{f(G)\}^2\\
    &=\mathbb{E}\left\{g(G)\right\}-\mathbb{E}\{f(G)\}^2\\
\end{align*}$$
Avec $g(x)=e^{-\lambda x+\frac{\lambda^2}{2}}f^2(x)$ 
\end{proof}


**Proposition :** *la dérivée peut s'écrire sous les deux formes suivantes*
$$
   \frac{\partial}{\partial \lambda}Var\left(X_\lambda\right)
   =  \mathbb{E}\left\{(\lambda - G)e^{-\lambda G + \frac{\lambda^2}{2}} f^2(G)\right\}
   =  - \mathbb{E}\left\{ G e^{-2\lambda G - \lambda^2} f^2(G+\lambda)\right\}.  
$$


On a 
$$\begin{align*}
     \frac{\partial}{\partial \lambda}Var\left(X_\lambda\right)&= \frac{\partial}{\partial \lambda}\mathbb{E}\left\{e^{-\lambda G+\frac{\lambda^2}{2}}f^2(G)\right\}\\
     &=\mathbb{E}\left\{(\lambda - G)e^{-\lambda G + \frac{\lambda^2}{2}} f^2(G)\right\}
\end{align*}$$
D'autre part, avec $h(x)=(\lambda - x)e^{-\lambda G + \frac{\lambda^2}{2}} f^2(x)$:
$$\begin{align*}
    \mathbb{E}\left\{(\lambda - G)e^{-\lambda G + \frac{\lambda^2}{2}} f^2(G)\right\}&= \mathbb{E}\left\{h(G)\right\}\\
    &=\mathbb{E}\left\{e^{-\lambda G - \frac{\lambda^2}{2}} h(G+\lambda)\right\}\\
    &= - \mathbb{E}\left\{ G e^{-2\lambda G - \lambda^2} f^2(G+\lambda)\right\}
\end{align*}$$

On vérifie ce fait par simulation en tracant les deux approximations
Monte-Carlo de la dérivées.
